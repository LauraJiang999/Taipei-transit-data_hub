```python
def get_data_from_BQ(client:bigquery.Client):
    query = client.query(
        """
        SELECT * FROM `ETL_FACT.FACT_bike_realtime`
        ORDER BY `source_time` DESC
        LIMIT 1000
        """
    )
    df = query.to_dataframe()
    return(df)
df = get_data_from_BQ(client=BQ_CLIENT)
```
