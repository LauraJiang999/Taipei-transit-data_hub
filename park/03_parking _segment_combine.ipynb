```python
import pandas as pd 
import numpy as np
```

```python
file_path1 = '/Users/taylorlin/Desktop/TIR101_Group2/01_parking_TransfromData_Taipei.csv'
file_path2 = '/Users/taylorlin/Desktop/TIR101_Group2/02_parking_TransfromData_NewTaipei.csv'
df_TPE = pd.read_csv(file_path1)
df_NWT = pd.read_csv(file_path2)
```

```python
# address 後面要加 lng, lat
# 確定 'address' 欄位的索引位置
index_address = df_NWT.columns.get_loc('address') + 1

# 在 'address' 欄位後面插入 'lng' 欄位，值為空（NaN）
df_NWT.insert(loc=index_address, column='lng', value=np.nan)

# 'lng' 欄位已經插入，所以新的插入位置是 'address' 位置後兩位
df_NWT.insert(loc=index_address + 1, column='lat', value=np.nan)
```

```python
df_NWT = df_NWT.iloc[:, 1:]
```

```python
df_TPE = df_TPE.iloc[:,1:]
```

```python
df_combine = pd.concat([df_TPE, df_NWT], ignore_index=True)
df_combine
```

```python
df_combine.to_csv("03_parking_segment_Data.csv")
```
