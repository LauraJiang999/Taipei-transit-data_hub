```python
sql = """
    select bike_station_id , station_name
    from bike_station;
"""
with create_engine(f"mysql+pymysql://{username}:{password}@{server}/{db_name}",).connect() as conn:
    df_sql = pd.read_sql(sql,conn)
```

```python
def get_station_id(station_name):
    # 假设你有一个包含站点 ID 和站点名称的字典
    stations = {
        '-1': '對應不到bike站名',
        '500101001': '捷運科技大樓站'
        # 在这里添加更多的映射关系
    }
    
    # 根据提供的站点名称查找对应的站点 ID
    for station_id, name in stations.items():
        if name == station_name:
            return station_id
    
    # 如果找不到对应的站点名称，返回空字符串
    return ''

# 测试函数
station_name = '捷運科技大樓站'
station_id = get_station_id(station_name)
print(station_id)
```

```python
mapping = pd.Series(df_sql['bike_station_id'].values,index=df_sql['station_name']).to_dict()
def get_station_id(station_name):
    
    return mapping.get(station_name,-1)
```

```python
df['lend_station_name'] = df['lend_station_name'].apply(get_station_id)
```
