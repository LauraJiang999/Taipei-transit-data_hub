##### 方法一
- 從open data  download file (json)
- 將download file (json) 寫入ＭongoDB

```python
# 連線ＭongoDB ,建立db ＆collection
import json
import pymongo
myclient = pymongo.MongoClient("mongodb://root:password@localhost:27000")
mydb = myclient["traffic"]
mycol = mydb["bike"]
```

```python
# 確認連線的ＭongoDB資訊
myclient.server_info()
```

```python
# 從json檔資料讀出來
# 讀出來的json 暫存在 data
with open('bike_realtime_info2.json', 'r') as f:
    data = json.load(f)
```

```python
# 確認上一步驟的資料類型
print(type(data))
print(type(data[0]))
```

```python
#暫存在 data的資料 insert到MongoDB
mycol.insert_many(data)
```
