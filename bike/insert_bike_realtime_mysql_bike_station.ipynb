```python
# index_col=0 不要產生預設的index
import pandas as pd
path="/Users/laura/Desktop/TIR101_Group2/bike/bile_usage_realtime.csv"
df = pd.read_csv(path,index_col=0)
```

```python
# CREATE TABLE `bike_station` (
#   `bike_station_id` int PRIMARY KEY,
#   `station_name` varchar(100),
#   `total_space` int,
#   `lat` decimal(8,6),
#   `lng` decimal(9,6),
#   `city_code` char(3),
#   `district` varchar(20),
#   `address` varchar(100),
#   `disable` boolean,
#   `create_time` timestamp,
#   `update_time` timestamp
# );
```

```python
df.head()
```

```python
#刪除不要的欄位
df = df.drop(['sbi','mday','sareaen','snaen',
              'aren','srcUpdateTime','infoTime',
              'infoDate','bemp','updateTime'
              
              ],axis=1,errors='ignore')
df.head()
```

```python
# 使用mapping 重新定義欄位新名稱
mapping = {'sno':'bike_station_id',
           'sna':'station_name',
           'tot':'total_space',
           'sarea':'district',
           'lat':'lat',
           'lng':'lng',
           'ar':'address',
           'act':'disable'
           }
```

```python
# 將重新定義的新欄位名稱 套用到df 
# ,axis=1 ->指定‘欄’
df = df.rename(mapping,axis=1)
```

```python
# 新增欄位'city_code'
# 將欄位'city_code' 給統一值'TPE'
df['city_code']='TPE'
```

```python
df
```

```python
# 單獨確認->處理將'YouBike2.0_'移除
text = 'YouBike2.0_捷運科技大樓站'
prefix = 'YouBike2.0_'
result = text.replace(prefix, '')
```

```python
# 處理將'YouBike2.0_'移除 ->定義為函式
def rename_station(text):
    prefix = 'YouBike2.0_'
    result = text.replace(prefix, '')
    return result
```

```python
# 處理將'YouBike2.0_'移除的函式 套用到dataframe
df['station_name'] = df['station_name'].apply(rename_station)
```
