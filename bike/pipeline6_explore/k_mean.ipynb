```python
import pandas as pd
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('bike_station_grouping_features.csv')
```

```python
data
```

```python
tpe_stations = pd.read_csv('TPE_bike_station.csv')
```

```python
data.index=data['station']
result = pd.DataFrame(data.index)
data.drop(['station_no','station'],axis=1,inplace=True)
```

```python
cat_col = [col for col in data.columns if 'hour_' in col] #this needs to be one hot if included
```

```python
data.drop(cat_col,axis=1,inplace=True)
```

```python
scaler = StandardScaler()
```

```python
data = pd.DataFrame(scaler.fit_transform(data),columns=data.columns,index=data.index)
```

```python
k_means_5 = KMeans(n_clusters=5)
k_means_10 = KMeans(n_clusters=10)
```

```python
k_means_5.fit(data)
k_means_10.fit(data)
```

```python
result['kmeans_5'] = k_means_5.predict(data)
result['kmeans_10'] = k_means_10.predict(data)
```

```python
DBSCAN_euclidean_model = DBSCAN(eps=1.5,min_samples=5,metric='euclidean')
result['DBSCAN_euclid_1point5'] = DBSCAN_euclidean_model.fit_predict(data)
DBSCAN_cosine_model = DBSCAN(eps=0.1,min_samples=5,metric='cosine')
result['DBSCAN_cosine_0point1'] = DBSCAN_cosine_model.fit_predict(data)
```

```python
tpe_stations = tpe_stations.merge(result,how='inner',left_on='name_tw',right_on='station')
```

```python
tpe_stations.to_csv('cluster_explore_first_try.csv',index=False)
```
